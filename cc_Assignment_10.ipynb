```python
texts = [
    "The camera quality of this smartphone is amazing and crystal clear.",
    "Battery life is impressive, lasting more than two days with heavy use.",
    "Build quality feels premium, and the display is vibrant and responsive."
]
```

```python
from sklearn.feature_extraction.text import CountVectorizer

count_vectorizer = CountVectorizer()
bow_matrix = count_vectorizer.fit_transform(texts)

# Convert to DataFrame for readability
import pandas as pd
bow_df = pd.DataFrame(bow_matrix.toarray(), columns=count_vectorizer.get_feature_names_out())
print("Bag of Words Matrix:")
print(bow_df)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(texts)
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf_vectorizer.get_feature_names_out())
print("\nTF-IDF Matrix:")
print(tfidf_df)
```

```python
import numpy as np

feature_names = tfidf_vectorizer.get_feature_names_out()
print("\nTop 3 keywords from each text:")
for i, row in enumerate(tfidf_matrix.toarray()):
    top_indices = np.argsort(row)[::-1][:3]  # Top 3 scores
    top_keywords = [feature_names[j] for j in top_indices]
    print(f"Text {i+1}: {top_keywords}")
```

```python
text1 = """Artificial Intelligence allows machines to mimic human intelligence.
It powers applications like image recognition, voice assistants, and autonomous vehicles.
AI systems learn from data and improve performance over time."""

text2 = """Blockchain is a decentralized technology used for secure data transactions.
It ensures transparency, immutability, and trust in digital ledgers.
Blockchain powers cryptocurrencies and supply chain tracking."""
```

```python
import re

def tokenize(text):
    # Lowercase and remove non-alphabetic characters
    tokens = re.findall(r'\b[a-zA-Z]+\b', text.lower())
    return set(tokens)

tokens1 = tokenize(text1)
tokens2 = tokenize(text2)

print("Tokens (AI):", tokens1)
print("Tokens (Blockchain):", tokens2)
```

```python
def jaccard_similarity(set1, set2):
    intersection = set1.intersection(set2)
    union = set1.union(set2)
    return len(intersection) / len(union)

jac_sim = jaccard_similarity(tokens1, tokens2)
print(f"\n🔹 Jaccard Similarity: {jac_sim:.4f}")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform([text1, text2])

cos_sim = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:2])
print(f"🔹 Cosine Similarity: {cos_sim[0][0]:.4f}")
```

```python
reviews = [
    "The delivery was super fast and the packaging was excellent!",
    "The product quality is terrible and I’m very disappointed.",
    "Customer service was okay, but could be better.",
    "I love this! Works exactly as described and looks great.",
    "Not worth the money. It stopped working after two days."
]
```

```python
from textblob import TextBlob

def analyze_sentiment(review):
    blob = TextBlob(review)
    polarity = blob.sentiment.polarity
    subjectivity = blob.sentiment.subjectivity

    if polarity > 0.1:
        sentiment = 'Positive'
    elif polarity < -0.1:
        sentiment = 'Negative'
    else:
        sentiment = 'Neutral'

    return polarity, subjectivity, sentiment

results = [analyze_sentiment(r) for r in reviews]
```

```python
import pandas as pd

df = pd.DataFrame(reviews, columns=["Review"])
df[['Polarity', 'Subjectivity', 'Sentiment']] = pd.DataFrame(results)
print(df)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Combine all positive reviews
positive_text = " ".join(df[df["Sentiment"] == "Positive"]["Review"])

wordcloud = WordCloud(width=800, height=400, background_color='white').generate(positive_text)

# Plot it
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title("Word Cloud of Positive Reviews", fontsize=14)
plt.show()
```

```python
text = """Artificial intelligence is changing the world. From self-driving cars to virtual assistants,
AI is integrated into our daily lives. It helps make smarter decisions, automates repetitive tasks,
and enables machines to learn from data. As AI continues to evolve, it promises to revolutionize industries,
improve healthcare, and enhance education. However, ethical concerns about bias and privacy remain critical.
Responsible development is key to ensuring that AI benefits society. With continuous innovation and proper regulation,
AI has the potential to become one of the most transformative technologies of our time."""
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
import numpy as np

# Tokenize the text
tokenizer = Tokenizer()
tokenizer.fit_on_texts([text])
total_words = len(tokenizer.word_index) + 1

# Convert to sequences
input_sequences = []
for line in text.split('.'):
    tokens = tokenizer.texts_to_sequences([line])[0]
    for i in range(1, len(tokens)):
        n_gram_seq = tokens[:i+1]
        input_sequences.append(n_gram_seq)
```

```python
max_seq_len = max(len(seq) for seq in input_sequences)
input_sequences = pad_sequences(input_sequences, maxlen=max_seq_len, padding='pre')

# Split into features and labels
X = input_sequences[:, :-1]
y = input_sequences[:, -1]

# One-hot encode the labels
from keras.utils import to_categorical
y = to_categorical(y, num_classes=total_words)
```

```python
from keras.models import Sequential
from keras.layers import Embedding, LSTM, Dense

model = Sequential()
model.add(Embedding(input_dim=total_words, output_dim=10, input_length=max_seq_len-1))
model.add(LSTM(100))
model.add(Dense(total_words, activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X, y, epochs=200, verbose=0)
```

```python
def generate_text(seed_text, next_words=10):
    for _ in range(next_words):
        token_list = tokenizer.texts_to_sequences([seed_text])[0]
        token_list = pad_sequences([token_list], maxlen=max_seq_len-1, padding='pre')
        predicted = model.predict(token_list, verbose=0)
        predicted_word = tokenizer.index_word[np.argmax(predicted)]
        seed_text += ' ' + predicted_word
    return seed_text

# Example:
print(generate_text("artificial", next_words=10))
```
