## Q.6 Create a sample dataset (employees.csv) containing information about employees in a company. 
.

```python
data2 = {
    "Employee_ID" : [101, 102, 103, 104, 105],
    "Name" : ["Alice", "Bob", "Charlie", "Diana", "Edward"],
    "Department" : ["HR", "IT", "IT", "Marketing", "Sales"],
    "Age" : [29, 34, 41, 28, 38],
    "Salary" : [50000, 70000, 65000, 55000, 60000],
    "Years_of_Experience" : [4, 8, 10, 3, 12],
    "Joining_Date" : ["2020-03-15", "2017-07-19", "2013-06-01", "2021-02-10", "2010-11-25"],
    "Gender" : ["Female", "Male", "Male", "Female", "Male"],
    "Bonus" : [5000, 7000, 6000, 4500, 5000],
    "Rating" : [4.5, 4.0, 3.8, 4.7, 3.5]
}
df2 = pd.DataFrame(data2)
print(df2)
```

#### 5. Calculate the following statistics from the dataset:
- The average salary of employees
- The total bonus paid to all employees
- The youngest employee's age
- The highest performance rating

```python
print("Average Salary:", df2["Salary"].mean())
print("Total Bonus:", df2["Bonus"].sum())
print("Youngest Employee Age:", df2["Age"].min())
print("Highest Performane Rating:", df2["Rating"].max())
```

#### 6. Sort the DataFrame by the Salary column in descending order.

```python
sorted_df2 = df2.sort_values(by="Salary", ascending=False)
print(sorted_df2)
```

#### 7. Add a new column that categorizes employees based on their performance rating:
- Excellent for ratings >= 4.5
- Good for ratings >= 4.0 but < 4.5
- Average for ratings < 4.0

```python
df2["Performance"] = pd.cut(
    df2["Rating"],
    bins = [0, 4.0, 4.5, 5],
    labels = ["Average", "Good", "Excellent"]
)
print(df2)
```

#### 8. Identify missing values in the DataFrame.

```python
print(df2.isnull().sum())
```

#### 9. Rename the Employee_ID column to ID.

```python
df2 = df2.rename(columns={'Employee_ID':'ID'})
print(df2)
```

#### 10. Find all employees who:
- Have more than 5 years of experience.
- Belong to the IT department.

```python
print(df2[df2['Years_of_Experience'] > 5])
print(df2[df2['Department'] == 'IT'])
```

#### 11. Modify the dataset by adding a new column, Tax, which deducts 10% of the Salary.

```python
df2["Tax"] = df2["Salary"] * 0.1
print(df2)
```
