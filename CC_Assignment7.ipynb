```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Part I: Randomized Sales Data Generation

```python
roll_number = 102317192
np.random.seed(roll_number)
```

## Generate sales data (12 months, 4 categories)

```python
sales_data = np.random.randint(1000, 5001, size = (12, 4))
```

## Convert to dataframe

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'June', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
categories = ['Electronics', 'Clothing', 'Home &  Kitchen', 'Sports']
df = pd.DataFrame(sales_data, columns = categories, index = months)
```

## Display first 5 rows and summary statistics

```python
print(df.head())
print(df.describe())
```

## Total sales per category and per month

```python
df['Total Sales'] = df.sum(axis = 1)
total_sales_per_category = df[categories].sum()
total_sales_per_month = df['Total Sales']
```

## Average sales growth between consecutive months

```python
df['Growth Rate'] = df['Total Sales'].pct_change() * 100
```

## Apply a randomized discount

```python
if roll_number % 2 == 0:
    df['Electronics'] *= 0.9
else:
    df['Clothong'] *= 0.85
```

# Part III: Visualizations

```python
plt.figure(figsize = (10, 6))
for category in categories:
    plt.plot(df.index, df[category], marker = 'o', label = category)
plt.xlabel("Months")
plt.ylabel("Sales")
plt.title("Monthly Sales Trends")
plt.legend()
plt.show()
```

## Box plot

```python
plt.figure(figsize = (8, 5))
sns.boxplot(data = df[categories])
plt.title("Sales Distribution by Category")
plt.show()
```

# Q4: Swap elements in a list

```python
def swap_list_elements(lst, i, j):
    temp = lst[i]
    lst[i] = lst[j]
    lst[j] = temp
    return lst
sample_list = [1, 2, 3,4,5]
swap_list_elements(sample_list, 1 , 3)
```

# Q4: Swap elements in a set by converting it to a list

```python
def swap_set_elements(s, i, j):
    s_list = list(s)
    s_list[i], s_list[j] = s_list[j], s_list[i]
    return set(s_list)
sample_set = {10, 20, 30, 40, 50}
swap_set_elements(sample_set, 1,3)
```
