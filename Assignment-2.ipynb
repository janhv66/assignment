## 1. Create a List L that is defined as= [10, 20, 30, 40, 50, 60, 70, 80]. 
1. WAP to add 200 and 300 to L. 
2. WAP to remove 10 and 30 from L. 
3. WAP to sort L in ascending order. 
4. WAP to sort L in descending order. 

```python
L = [10, 20, 30, 40, 50, 60, 70, 80]
L.append(200)
print("List after adding 200:", L)
L.append(300)
print("List after adding 300:", L)
L.remove(10)
print("List after removing 10:", L)
L.remove(30)
print("List after removing 30:", L)
L.sort()
print("List after sorting in ascending order:", L)
L.sort(reverse = True)
print("List after sorting in descending order:", L)
```

## 2. Create a tuple of marks scored as scores = (45, 89.5, 76, 45.4, 89, 92, 58, 45) and perform the following operations using tuple functions: 
1. Identify the highest score and its index in the tuple. 
2. Find the lowest score and count how many times it appears. 
3. Reverse the tuple and return it as a list. 
4. Check if a specific score ‚Äò76‚Äô (input by the user) is present in the tuple and print its first occurrence index, or a message saying it‚Äôs not present.

```python
scores = (45, 89.5, 76, 45.4, 89, 92, 58, 45)
highest_score = max(scores)
ind = scores.index(highest_score)
print(f"{highest_score} is the highest score at index {ind}")
lowest_score = min(scores)
count_lowest = scores.count(lowest_score)
print(f"{lowest_score} is the lowest score which occurs {count_lowest} times")
scores = scores[::-1]
print("The tuple after reversing:", list(scores))
score = int(input("Enter the score:"))
if score in scores:
    print(f"{score} occurs at index {scores.index(score)} in scores")
else:
    print(f"{score} is not present in scores")
```

## 3. WAP to create a list of 100 random numbers between 100 and 900. Count and print the:
1. All odd numbers 
2. All even numbers 
3. All prime numbers 

```python
import random
import math
def isPrime(n):
    for i in range(2, int(math.sqrt(n))):
        if n % i == 0:
            return 1
    return 0
l = [random.randint(100, 900) for i in range(100)]
odd = [num for num in l if num % 2 != 0]
even = [num for num in l if num % 2 == 0]
prime = [num for num in l if isPrime(num)]
print("Odd Numbers:")
for i in odd:
    print(i, end = " ")
print("\nEven Numbers:")
for i in even:
    print(i, end = " ")
print("\nPrime Numbers:")
for i in prime:
    print(i, end = " ")
```

## 4. Consider the following two sets, A and B, represen ng scores of two teams in multiple matches.  A = {34, 56, 78, 90}  and B = {78, 45, 90, 23} 
## WAP to perform the following opera ons using set func ons: 
1. Find the unique scores achieved by both teams (union of sets). 
2. Identify the scores that are common to both teams (intersec on of sets). 
3. Find the scores that are exclusive to each team (symmetric difference). 
4. Check if the scores of team A are a subset of team B, and if team B's scores are a superset of team A. 
5. Remove a specific score ùëã (input by the user) from set A if it exists. If not, print a message saying it is not present. 

```python
A = {34, 56, 78, 90}
B = {78, 45, 90, 23}
print("The unique scores achieved by both teams are:", A.union(B))
print("The scores that are common to both teams:", A.intersection(B))
print("The scores that are exclusive to each team:", A^B)
if A.issubset(B):
    print("Scores of team A are a subset of team B")
else:
    print("Scores of team A are not a subset of team B")
if B.issuperset(B):
    print("Scores of team B are a superset of team A")
else:
    print("Scores of team B are not a superset of team A")
score = int(input("Enter the score to be removed:"))
if score in A:
    A.remove(score)
else:
    print(f"{score} is not present in A")
```

## 5. Write a program to rename a key city to a location in the following dictionary.
